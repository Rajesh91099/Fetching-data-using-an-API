# Creating a dataset of top-rated movies from TMDB website using API

```python
#importing required libraries to fetch data from an API
import pandas as pd
import requests
```

```python
#By using requests library we will fetch data in json format and store it inside response variable
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=538554910b982fa866e570df30493201&language=en-US&page=1')
```

```python
#Now we will see the json text
response.json()
```

```python
#Here we only need results in json dictionary so we fetch just results 
response.json()['results']
```

```python
#We will select the required columns for our dataframe and store it in tempdf variable
tempdf = pd.DataFrame(response.json()['results'])[['id','title','original_language','release_date','popularity','vote_average','vote_count']]
```

```python
#lets call tempdf to see our dataframe
tempdf
```

```python
#As of now we just data from a single page, so we are going to create a loop to fetch data from all pages
#First we will create an empty dataframe
df = pd.DataFrame()
```

```python
for i in range(1,570):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=538554910b982fa866e570df30493201&language=en-US&page={}'.format(i))
    tempdf = pd.DataFrame(response.json()['results'])[['id','title','original_language','release_date','popularity','vote_average','vote_count']]
    df = df.append(tempdf,ignore_index=True)
```
